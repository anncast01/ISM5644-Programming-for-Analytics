#### 3. Filtering
* Display the name, sugar level, and rating of cereals that contain more than 10 grams of sugar

```python
high_sugar_cereals = df.loc[df['sugars'] > 10, ['name', 'sugars', 'rating']]
high_sugar_cereals
```

```python
avg_rating_HighSugar = df.loc[df['sugars'] > 10, 'rating'].mean()
avg_rating_LowSugar = df.loc[df['sugars'] < 10, 'rating'].mean()
avg_rating_HighSugar
avg_rating_LowSugar
```

```python
high_cal_cereals = df.loc[df['calories'] > df['calories'].mean(), ['name', 'manufacturer', 'rating']]
high_cal_cereals
```

```python
df = df.loc[df['manufacturer'] != 'R', :]
df
```

```python
gm_above_AvgRating = df.loc[(df['manufacturer'] == 'G') & (df['rating'] > df['rating'].mean()), :]
gm_above_AvgRating
```

#### 4. Grouping
* Group data by the manufacturer and find the average sugar level and the average rating for each group
* Select the observations with high rating (rating > 50) and find out how many were manufactured by each manufacturer.

```python
manudata_avg = df.groupby('manufacturer')[['sugars', 'rating']].mean()
manudata_avg
```

```python
highRating = df.loc[df['rating'] > 50, 'manufacturer'].value_counts()
highRating
```
