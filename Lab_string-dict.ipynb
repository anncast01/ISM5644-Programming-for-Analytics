#### 1. Assume that there exists two lists in memory that are of the same length. The first list contains user names (e.g., ['Jane','Joe','Sally']) and the second list contains user IDs (e.g., [101,102,103]). Write a program that creates a single dictionary which maps user names to user IDs. In the dictionary, names should be used as keys and IDs should be used as values. The user name list can be of arbitrary length.
Hint: Use a loop to iterate over the length of lists and obtain values using indices.

```python
a_list = ['Jane','Joe','Sally']
user_ids = [101,102,103]
newDict = {}
for i in range(len(a_list)):
    value = user_ids[i]
    key = a_list[i]
    newDict[key]=value
print(newDict)
```

#### 2. Assume that there exists a dictionary that stores integer values. Write a program that asks the user for an integer input, searches this input in the <u>values</u> of the dictionary, and prints out the input's associated key. If the input is not in the dictionary, the program should print out a <u>not found</u> message.
Hint: To test the program, you can initialize an arbitrary dictionary such as d = {'a':10,'b':20,'c':30}

```python
d = {'a':10,'b':20,'c':30}
x = int(input('Enter an integer:'))
matchList = False 
for key,value in d.items():
    if value == x:
        matchList = True
        break 
if matchList: 
    print ('The associated key for your value is:', key)
else:
    print('not found')
```

#### 6. Write a program that prompts the user for a number input with 10 digits (such as 8005551212). If the input is 10 digits, the program should divide this input into three parts consisting of 3 digits, 3 digits, and 4 digits and print them as a single string joined by hyphens (-). For example, if the input is 8005551212, then the output should be: “800-555-1212”. If the user input is <u>not</u> in 10 digits, it should display a warning message and prompt the user again. The program should run <u>iteratively</u> until a 10 digit input is entered.
Hint: You can initialize the program with a While loop that will iteratively run until a condition has been met. 

```python
while True:
    number = input('Enter a 10-digits number:')
    if len(number) == 10:
        str_number1 = number[0:3]
        str_number2 = number[3:6]
        str_number3 = number[6:]
        print(str_number1,'-',str_number2,'-',str_number3)
        break 
    else:
        print('Not 10 digits. Try again.')
```

#### 7. #### 8. Given a person’s full name as a string, write an expression that prints the the initials of the person -- each character followed by a period. For example, if the string is ‘Jane Doe’, then it should print ‘J.D.’. If the person's name is 'Jane Mary Berry Doe', then it should print 'J.M.B.D.' Note that the middle name is optional; there can be zero or more middle names included in the full name. Also, the length of the first / middle / last names can vary.
Hint: You can use the .split() method to spearate the individual words in the name string.

```python
FullName = 'Jane Doe'
initials = ''
list_FullName = FullName.split(' ')
for i in range(len(list_FullName)):
    initials += list_FullName[i][0]+'.'
    
print('Your initials are:', initials)
```

#### 9. Convert a string that consists of two words that are both capitalized into a string where only the first letters of the words are capitalized. For example, convert “NEW YORK” to “New York”.
Note: You are <u>*not*</u> allowed to use the .title() method.

```python
st = 'NEW YORK'
list_st = st.split(' ')
finalWord = ' '
for word in list_st:
    finalWord += word.capitalize()+ ' '
print(finalWord)
```

#### 10. Write a program that returns True if a given string (user input) includes every letter of the alphabet. If the string does not contain each letter, print which ones are missing. The results should <u>not</u> depend on capitalization.
Hint: You can use the following code snippet to load the letters in the alphabet into a string called alphabet_str: <br>
*import string <br>
alphabet_str = string.ascii_lowercase* <br>
Next, you can loop over the alphabet letters to check if they are included in the string or not.

```python
import string
alphabet_str = string.ascii_lowercase
letters = input('Enter a word:').lower()
missing_letters = ''
for character in alphabet_str:
    if letters.find(character) == -1:
        missing_letters += character
if missing_letters: 
    print('This letters are missing:', missing_letters)
else: 
    print('All alphabet letters included!')
```
