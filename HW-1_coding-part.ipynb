#### 6. Given a list of numbers, create a new list of numbers such that the first and last numbers are added and stored as the first number, the second and second-to-last numbers are stored as the second number, and so on. Note that you need to check for even and odd length of lists. In case of an odd number of integers, add the central integer in the original list to the end of the new list.

```python
numbers = [1,2,3,4,5]
new_list = []
a = 0 
b =len(numbers)-1
while a<b:
    total = numbers[a] + numbers[b]
    new_list += [total]
    a += 1
    b -= 1
if a == b:
    new_list += [numbers[a]]
print(new_list)
```

#### 8. Write a program that finds all the perfect numbers between 1 and *n*, where *n* is a user input collected by the program. 
#### *Definition*: A perfect number is a positive integer that is equal to the sum of its positive factors, excluding the number itself. For instance, the number 6 has the factors: 1, 2, and 3. The sum of the factors are: 1+2+3 = 6, which is the number itself. Accordingly, the number 6 is a perfect number.
Hint: You can use / expand the code from an earlier lab question (i.e., finding all the factors of a number) to answer this question. 

```python
n = int(input('Enter a number (limit):'))
for x in range (1,n+1):
    numFactor = [n for n in range(1, x) if x % n == 0]
    if sum(numFactor) == x:
        print(x,'is a perfect number.')
```

#### 10. Write a program that prompts for an integer - let's call it X - and then finds the consecutive sums of all integers between 1 and X. For example, suppose that 5 is entered, i.e., X = 5. Then, you will need to obtain five sums of consecutive numbers as:
1 = 1 <br>
1+2 = 3 <br>
1+2+3 = 6 <br>
1+2+3+4 = 10 <br>
1+2+3+4+5 = 15 <br>
#### Ideally, the output of the program should be as it is shown above. If you cannot get the left hand sides printed out (i.e., individual summation elements and '+' signs), then it is OK to only print out the total sums (1, 3, 6, 10, and 15 for the example above) on separate lines.

```python
x = int(input('Enter an interger:'))
#first for controls the amount of lines being printed
for n in range (1,x+1):
    total = 0
    sequence = ''
    #second for is for the sum of the current line
    for m in range(1,n+1):
        total += m
        sequence += str(m)
        if m < n:
            sequence += '+'
    print(sequence,'=',total)
```
