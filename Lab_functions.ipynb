#### 1. Write a function called `myCalc` that receives two int arguments, x1 and x2. The function returns their sum, if x1 > x2. It returns their product if x1 <= x2.

```python
def summation(x,y):
    return x+y
def multiplication(x,y):
    return x*y
    
def myCalc():
    x1 =int(input('First number: '))
    x2 = int(input('Second number: '))

    if x1 > x2:
        output = summation(x1,x2)
    elif x1 <= x2:
        output = multiplication(x1,x2)
        
    print(output)
```

#### 2. Write a function called `primeCheck` that receives an integer argument (user input) and returns True if the argument is a prime number and returns False otherwise.
Hint: You can re-use / modify the prime number checking code that you had worked on an earlier lab to answer this question.

```python
def primeCheck(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False          
    return True
            
userInput = int(input('Enter a number: '))
output = primeCheck(userInput)
print(output)
```

#### 3. Write a program to calculate and print the sum of all prime numbers between 2 and *N*, where *N* is a user input. 
Note: For this question, you have to iteratively invoke the primeCheck function that you have created in the previous question.

```python
def primeCheck(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

x = int(input('Enter a number: '))
sumPrime = 0
for num in range(2, x + 1):
    if primeCheck(num):
        sumPrime += num 
print("The sum of all prime numbers is:", sumPrime)
```

#### 4. Write a program that prompts the user for a sentence with multiple words in it. Words can be lowercase, uppercase, or mixed. The program prints out each word in the sentence with the first letter capitalized and the rest being lowercase. 
#### For transforming the words, <u>use a function</u>  and invoke the same function for each word in the sentence. For example, the user input: `thIS question AGAIN??` should be printed out as `This Question Again??`.

```python
def capt(x):
    if len(x) == 0:
        return x
    return x[0].upper() + x[1:].lower()

userInput = input('Enter a sentence: ')
xList = userInput.split(' ')
newWords= []
    
for i in xList:
    newWords.append(capt(i))
        
final = ' '.join(newWords)
print(final)
```

#### 5. Write a program that prompts the user for a sentence with multiple words in it. The program prints out the sentence with every single vowel (a, e, i, o, u) replaced by an asterisk (*), regardless of whether the vowel is uppercase or lowercase. 

#### For transforming the words, use a function (e.g., `censor_word)` and invoke the same function for each word in the sentence before printing the final result.

```python
def censor_word(word):
    vowels = "aeiouAEIOU"
    censored = ""
    for i in word:
        if i in vowels:
            censored += "*"
        else:
            censored += i
    return censored

userInput = input("Enter a sentence: ")
word_list = userInput.split(" ")
censoredSentence = []
    
for word in word_list:
    censoredSentence.append(censor_word(word))
        
final_sentence = " ".join(censoredSentence)
print(final_sentence)
```

#### 6. A retail store is offering 10% discount for all its customers. Customers who are members of this store get a bonus 10% discount (i.e., 20% discount in total). Write a function that takes as input the retail cost of the item that a person is purchasing and a Boolean membership variable (indicating whether the person is a member or not). The function should return the final discounted value of the item. For example, if Person A is a member and the retail cost of the item is 100 USD, then the final discounted value should be 80 USD.

```python
def calculate_discount(price, member):
    if member:
        discount = 0.20
    else:
        discount = 0.10
        
    final_price = price * (1 - discount)
    return final_price


price = float(input("Enter the retail cost: "))
member_input = input("Is the customer a member? (yes/no): ").lower()
member = member_input == "yes"
    
result = calculate_discount(price, member)
print("The final discounted value is:", result)
```

#### 7. A palindrome is a word that is the same backwards as forwards. An example is the word “rotor”. First, define a function that can identify whether a word is a palindrome or not. Then, run this function on each element in the following list and find out how many words were palindromes.
Hint: To check if a word is a palindrome, you can reverse the word and do a Boolean comparison (==) with the original word.
#### possible_pal_list = [‘level’, ‘dugged’, ‘reviver’, ‘civil’, ‘civic’]

```python
def palindromeCheck(word):
    reversed_word = word[::-1]
    return word == reversed_word

possible_pal_list = ['level', 'dugged', 'reviver', 'civil', 'civic']
palin_count = 0
for word in possible_pal_list:
    if palindromeCheck(word):
        palin_count += 1
            
print("Number of palindromes found:", palin_count)
```

#### 8. A financial advisor wants a reusable tool to show clients how fast their money can grow using compound interest. Write a program that includes <u>two</u> functions to achieve this as follows:

1. The first function is called `calculate_years_to_goal`. It takes three parameters: `start_balance` (float), `interest_rate` (float, as a decimal), and `target_balance` (float).
    - Inside the function, use a `while` loop to simulate the passage of time year by year.
    - Each year, increase the balance by the interest rate (i.e., Balance = Balance + (Balance * Rate)).
    - The function should return the total number of years it took to reach or exceed the target balance.

2. The second function is the `main` function. It prompts the user for their starting amount, interest rate, and goal. It should then invoke your `calculate_years_to_goal` function with those inputs and print the resulting number of years.


```python
def calculate_years_to_goal(start_balance, interest_rate, target_balance):
    currentBalance = start_balance
    years = 0
    
    while currentBalance < target_balance:
        currentBalance = currentBalance + (currentBalance * interest_rate)
        years += 1
    return years

def main():
    start_amount = float(input("Enter starting amount: "))
    rate = float(input("Enter interest rate (decimal): "))
    goal = float(input("Enter goal amount: "))
    total_years = calculate_years_to_goal(start_amount, rate, goal)
    
    print("Total years to reach the goal:", total_years)
```
