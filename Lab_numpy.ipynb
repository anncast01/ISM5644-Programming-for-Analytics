#### 4. Please complete the steps below.
* Initialize two objects: 1) a 10x2 numpy array filled with zeros, 2) a list of integers between 10 and 20 (excluding 20).
* Iterate over the list (for loop) and modify the values in the array as follows:
  * If the list element is an even number, assign value 1 to the 1st column of the corresponding index in the array
  * If the list element is an odd number, assign value 1 to the 2nd column of the corresponding index in the array
* For instance, the first element in the list is 10, an even number, therefore array[0,0] should be 1.
* Print out the modified array.

```python
import numpy as np
arr = np.zeros((10, 2), dtype=int)
num_list = list(range(10, 20))
print(arr)
print(num_list)
print('###')
for i, num in enumerate(num_list):
    if num % 2 == 0:
        arr[i, 0] = 1
    else:
        arr[i, 1] = 1
print(arr)
```

#### 5. Please complete the steps below.
* Initialize a 1D array with the following integers: 15,12,2,43,12,6,6,15,8,2
* Some of the numbers above are repeated. Create a new array that includes all the numbers from the original array, but without any duplicates. The new array should preserve the order from the old one.
* Find the duplicated elements in the original array and print them out.

```python
import numpy as np
arr = np.array([15, 12, 2, 43, 12, 6, 6, 15, 8, 2])
print(arr)
print('###')
unique_ind = np.unique(arr, return_index=True)[1]
arr_no_dup = arr[np.sort(unique_ind)]
print(arr_no_dup)
print('###')
vals, counts = np.unique(arr, return_counts=True)
duplicates = vals[counts > 1]
print(duplicates)
```
