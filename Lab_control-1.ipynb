#### 5. Write a program that uses a "While" loop to identify and print every three-digit number that is divisible by 17. The program should also keep a count of identified numbers and print out the total count at the end of its execution.

```python
number = 100
total_count = 0 

while number <=999:
    if number % 17 ==0:
        print(number)
        total_count +=1
    number +=1
print('The total count is:', total_count)
```

#### 6. Write a program that prompts for an integer, called x, and then prints out the sum of x consecutive integers starting at 1. Your program should include a While loop.
Hint: Do not use 'sum' as a variable name in your program, as that is a special function in Python.

```python
x = int(input('Enter a number'))
counter = 1 
total = 0
while counter <= x:
    total += counter
    counter +=1
print(total)
```

#### 7. Write a program that continually prompts the user for positive integers and stops when the sum of numbers entered by the user exceeds 1,000. The program should:
* Use a while loop
* Print out the current cumulative sum at each iteration (even if it exceeds 1,000)
* Break early if a negative integer is entered
* Print out the final sum at the end

```python
total = 0
while total <= 1000:
    x = int(input('Enter a number greater than zero'))
    if x < 0: 
        break 
    total +=x 
    print('Current Cumulative Sum:', total)
print('Total Sum:', total)
```

#### 9.The Collatz conjecture is a conjecture in mathematics that concerns a sequence of positive numbers, ending at 1 using an algorithm. The algorithm is as follows:
* Ask the user for a numeric input
* If the number is odd, multiply by 3 and add 1
* If the number is even, divide by 2
* When the number reaches 1, quit
* At each iteration, print out the number on a new line
#### Implement this algorithm using while and if-else controls.

```python
x = int(input('Enter a number'))
while x!=1:
    if x % 2 != 0:
        x = (x*3)+1
    else:
        x = x // 2 
    print(x)
```

#### 10. Write a program that: 
* prompts the user for an integer input (greater than 2)
* checks to see if the input is prime.
* prints out the finding at the end of its execution (i.e., prime number or not a prime number).

```python
x = int(input('Enter a number greater than 2: '))
inputPrime = True
counter = 2

while counter < x:
    if x % counter == 0:
        inputPrime = False 
    counter +=1
if inputPrime == True:
    print('Your number is prime')
else:
    print('Your number is not prime')
```
