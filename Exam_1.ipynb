Yoda Speech generator: Given a sentence string that consists of multiple words, divide the sentence into two parts and re-order the parts into a new string. If the sentence contains an even number of words (e.g., 2n), then each part should contain the same number of words (e.g., n for each part). If the sentence contains an odd number of words, then the 1st part in the original sentence should contain one less word then the 2nd part in the original sentence (e.g., n-1 for the 1st part and n for the 2nd part). Here are some examples:

Original		Output (Yoda Speech)
Even	‘you must have patience’	=>	
‘have patience you must’

Even	
‘the future is always in motion’

=>	
‘always in motion the future is’

Odd	
‘this is just the beginning’

=>	
‘just the beginning this is’

Odd	
‘you have much to learn’

=>	
‘much to learn you have’

Optional: Feel free to add a ‘hmmmm..’ at the end of the new sentence.

#### Q4. Use the cell below to enter your code for this question.

```python
yoda = 'you must have patience'
new_yoda = ''
yoda_list = yoda.split(' ')
len_yoda = len(yoda_list)
a = 0
b = len_yoda - 1

while a<b: #if it's even amount of words
    #divide the sentence in two parts
    size = len_yoda // 2
    pt1 = str(yoda_list[0:size][::-1])
    pt2 = str(yoda_list[size:][::-1])
    a += 1
    b -= 1
    new_yoda += pt1 + pt2
if a == b: #if it's odd amount of words
    pt1 = str(yoda_list[0:size:-1])
    pt2 = str(yoda_list[size:])
    new_yoda += pt1 + pt2

print(new_yoda)
```
